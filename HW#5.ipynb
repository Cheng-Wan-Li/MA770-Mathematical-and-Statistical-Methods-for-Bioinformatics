```python
import numpy as np

def O(s):
    ### This function create the string O_t of observables.
    vec = np.zeros(len(s)+2)
    vec[len(s)+1]=5
    for n in range(0,len(s)):
        if s[n] == 'A':
            vec[n+1] = 1
        elif s[n] == 'C':
            vec[n+1] = 2
        elif s[n] == 'G':
            vec[n+1] = 3
        elif s[n] == 'T':
            vec[n+1] = 4
    return vec

P1 = np.array([
    [0,0.3,0.7,0,0,0],
    [0,0.7,0.3,0,0,0],
    [0,0,0,0.4,0.6,0],
    [0,0,0,0.5,0.5,0],
    [0.5,0,0,0,0,0.5],
    [0.7,0,0,0,0,0.3]
])
### This is our initial guess of the transition probability.
###'Begin' and 'End' states are added.

Q1 = np.array([
    [1,0,0,0,0],
    [0,0.7,0.1,0.1,0.1],
    [0,0.2,0.4,0.2,0.2],
    [0,0.1,0.1,0.7,0.1],
    [0,0.2,0.4,0.2,0.2],
    [0,0.1,0.1,0.1,0.7]
])
### This is our initial guess of the emission probability. 
###'Begin' and 'End' states are added.

def F(P,Q,O,i,t):
    ### This function calculate the forward probability.
    (n,m) = np.shape(Q1)
    if t > 0:
        ### The general case.
        return Q[i,np.int(O[t])]* sum(P[j,i]* F(P,Q,O,j,t-1) for j in range(0,n))
    else:
        ### The base case.
        return Q[i,np.int(O[0])]
        
def B(P,Q,O,i,t):
    ### This function calculate the backward probability.
    (n,m) = np.shape(Q1)
    l=len(O)-1
    if  t < l-1:
        ### The general case.
        return sum(P[i,j]* Q[j,np.int(O[t+1])]* B(P,Q,O,j,t+1) for j in range(0,n))
    elif t == l-1:
        ### The base case.
        return P[i,l-1]
    else:
        return 0

def FResults(P,Q,O):
    ### This function collect all the forward probabilities.
    (n,m) = np.shape(Q1)
    l=len(O)-1
    M=np.zeros((n,l))
    for i in range(0,n):
        for t in range(0,l):
            M[i,t]=F(P,Q,O,i,t)
    return M

def BResults(P,Q,O):
    ### This function collect all the forward probabilities.
    (n,m) = np.shape(Q1)
    l=len(O)-1
    M=np.zeros((n-1,l))
    for i in range(0,n-1):
        for t in range(0,l):
            M[i,t]=B(P,Q,O,i,t)
    return M
```

Let $M$ denote an arbitrary matrix with a complete orthonormal set of eigenvectors $\lambda_i$ and eigenvectors $\mathbf{v_i}$.

Show that
$$

\begin{align}

\begin{align}
M&=

\mathbf{v}_{1}\\

## Markov Chain: Reversibility
### Problem 6
Let $\{\psi(t)\}$ denote a Markov chain with possible underlying states $\{\varphi_1,\dots,\varphi_m\}$.

The initial distribution is selected to agree with the stationary distribution $\pi=(\pi_{k})$ which is given by,
$$
\pi_{k}=Prob(\psi(0)=\varphi_k)=\lim_{t\to\infty}Prob(\psi(t)=\varphi_k).
$$

The transition matrix is given by $P=(p_{ij})$ s.t.,
$$
p_{ij}=Prob(\psi(t+1)=\varphi_j|\psi(t)=\varphi_i)
$$

In addition we assume this chain to be irreducible and periodic.

Define the reverse sequence of $\{\psi(t)\}$ to be $\{\psi^{*}(t)\}$ s.t.,
$$
\psi^{*}(t)=\psi(T-t).
$$
### (a)
#### Theorem 1:
The prebabilistic structure of $\psi^{*}(t)$ is also regular Markov chain with transition matrix $P^{*}=(p^{*}_{ij})$ given by
$$
p^{*}_{ij}=p_{ji}\frac{\pi_{j}}{\pi_i}.
$$
Furthermore, the stationary distribution of reversed chain $\psi^{*}(t)$ is also $\pi$.

#### Proof:
By definition, we have
\begin{align}
p^{*}_{ij}&=Prob(\psi(t)=\varphi_j|\psi(t+1)=\varphi_i)\\
&=Prob(\psi(t+1)=\varphi_i|\psi(t)=\varphi_j)\frac{Prob(\psi(t)=\varphi_j)}{Prob(\psi(t+1)=\varphi_i)}\quad\text{(By Bayes' Theorem)}\\
&=p_{ji}\frac{\pi_{j}(t)}{\pi_{i}(t+1)}
\end{align}
By definition, we have:
$$
\lim_{t\to\infty}p_{ij}=p_{ij},
$$
and
$$
\lim_{t\to\infty}p^{*}_{ij}=p^{*}_{ij}.
$$
Therefore,
\begin{align}
p^{*}_{ij}
&=\lim_{t\to\infty}p^{*}_{ij}=p^{*}_{ij}\\
&=\lim_{t\to\infty}(p_{ji}\frac{\pi_{j}(t)}{\pi_{i}(t+1)})\\
&=p_{ji}\lim_{t\to\infty}\frac{\pi_{j}(t)}{\pi_{i}(t+1)}\\
&=p_{ji}\frac{\pi_{j}}{\pi_{i}}
\end{align}

Furthermore, the stationary distribution of the reversed chain is given by:
\begin{align}
\pi^{*}_j&=(\lim_{t\to\infty}\pi^{*}(0){P^{*}}^{t})_{j}\\
&=\lim_{t\to\infty}\sum_{i=1}^{n}\pi_{i}^{*}(0){P^{*}}^{t}_{ij}
\end{align}

Since
\begin{align}
p^{*}_{ij}p^{*}_{jk}&=p_{ji}\frac{\pi_{j}}{\pi_{i}}p_{kj}\frac{\pi_{k}}{\pi_{j}}\\
&=p_{kj}p_{ji}\frac{\pi_{k}}{\pi_{i}}
\end{align}
we can deduce that
\begin{align}
{P^{*}}^{t}_{ij}&={{P}^{t}}_{ji}\frac{\pi_{j}}{\pi_{i}}
\end{align}
by definition of matrix multiplication.

In addition, by construction of the reversed chain, we have:
\begin{align}
\pi_{i}^{*}(0)=\pi_{i}
\end{align}

Therefore, we have:
\begin{align}
\pi^{*}_j&=\lim_{t\to\infty}\sum_{i=1}^{n}\pi_{i}{{P}^{t}}_{ji}\frac{\pi_{j}}{\pi_{i}}\\
&=\lim_{t\to\infty}\sum_{i=1}^{n}\pi_{j}{{P}^{t}}_{ji}\\
&=\lim_{t\to\infty}[\pi{P}^{t}]_{j}\\
&=\pi_{j}
\end{align},
which verified our claim s,t, $\pi^*=\pi$.

### (b)
#### Proposition: 
If $\{\psi(t)\}$ is irreducible, then $\{\psi^{*}(t)\}$ is also irreducible.

#### Proof:
By hypothesis, we have:
\begin{align}
\forall i,j\ \exists k\ s.t.\ Prob(\psi_{t+k}=\varphi_j|\psi_t=\varphi_i)>0
\end{align}
Therefore, we must have $\pi_j>0$ for all $j$, otherwise, we cannot reach $\varphi_j$ in finite steps.

By construction, we also have:
\begin{align}
Prob(\psi_{t+k}=\varphi_j|\psi_t=\varphi_i)\pi_i=Prob(\psi_{t+k}=\varphi_j,\psi_t=\varphi_i)
\end{align}
i.e., $\{\psi_{t}\}$ is irreducible iff $Prob(\psi_{t+k}=\varphi_j,\psi_t=\varphi_i)>0$.

Then consider the reversed chain $\{\psi^{*}_{t}\}$ (without loss of generality, we assume the reversed chain start at $T=t+k$):

\begin{align}
Prob(\psi^{*}_{k}=\varphi_j,\psi^{*}_{0}=\varphi_i)&=Prob(\psi_{t}=\varphi_j,\psi_{t+k}=\varphi_i)\\
&=Prob(\psi_{t+k}=\varphi_j,\psi_{t}=\varphi_i)\\
&>0
\end{align}

Then, we have
\begin{align}
Prob(\psi^{*}_{k}=\varphi_j|\psi^{*}_0=\varphi_i)\pi^*_i=Prob(\psi^{*}_{k}=\varphi_j,\psi^{*}_{0}=\varphi_i)>0.
\end{align}
Since $\pi^*_i=\pi_i>0$, we can deduce that
$$
Prob(\psi^{*}_{k}=\varphi_j|\psi^{*}_0=\varphi_i)>0,
$$
which implies the reversed chain is irreducible.
